## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Creat blank figure 
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes 
axes.plot(x, y)
axes.set_xlabel('x') # Note the usr of set_ to begin the method
axes.set_ylabel('y')
axes.set_title('title')
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Create blank canvas
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,.5,.2,.2]) # insset axes

plt.show()
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Create blank canvas
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,0.5,0.2,0.2]) # insset axes

ax1.plot(x,y, 'r')
ax2.plot(x,y, 'r')

plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
# Create blank canvas
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,0.5,0.4,0.4]) # insset axes

plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Larger figure axes1 (main)
ax1.plot(x, z, 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


# Insert figure axes2 (inset)
ax2.plot(x, y,'b')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])

fig
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig,axes = plt.subplots(1,2) #(1,2) 1 row and 2 columns
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)

fig
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,axes = plt.subplots(1,2, figsize=(12,2)) #(1,2) 1 row and 2 columns 

axes[0].plot(x,y,color='blue',lw=3,ls='-')
axes[1].plot(x,z,color='red',lw=3,ls='--')

fig
```
